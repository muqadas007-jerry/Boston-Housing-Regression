```python
# Importing libraries
from os import X_OK
import pandas as pd
import numpy as np
import tensorflow as tf
import keras
from keras.utils import to_categorical,plot_model
from sklearn.pipeline import Pipeline as skPipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from imblearn.pipeline import Pipeline as imbPipeline
from sklearn.model_selection import GridSearchCV, RepeatedStratifiedKFold, train_test_split
import sklearn.metrics as mt
import seaborn as sns
import matplotlib.pyplot as plt

# Import Keras Wrapper connector
try:
    import scikeras
except ImportError:
    !python -m pip install scikeras

from scikeras.wrappers import KerasClassifier

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Importing original data from csv Obfuscated-MalMem2022.csv
original_data = pd.read_csv("/content/Obfuscated-MalMem.csv")
data = original_data.copy()
```

```python
# Malware categories
def classifyCategories(s):
  if s == "Benign":
    return 0  # Benign
  elif s.startswith("Spyware"): # Spyware
    return 1
  elif s.startswith("Ransomware"): # Ransomware
    return 2
  else: # Trojan Horse
    return 3

# Drop unnecesary column
data = data.drop(columns=["Class"])

# Classification data
y = data["Category"].map(lambda s: classifyCategories(s)).to_numpy()
x = data.drop(columns=["Category"]).to_numpy()

# Visualize general shapes on screen
print("Shape x (features):", x.shape)
print("Shape y (families labels): ", y.shape)

# Splitting (training and test balanced sets definitions)
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, shuffle=True, random_state = 1234, stratify = y)

# Number of classes
num_classes = 4

# ----------------------------------------------------------------------------
```

```python
# BAWS (Boundary-Aware and Weighted Synthesis) Implementation
# ----------------------------------------------------------------------------
class BAWS:
    def __init__(self, sampling_strategy='auto', k_neighbors=5, gamma=0.1, random_state=None):
        self.sampling_strategy = sampling_strategy
        self.k_neighbors = k_neighbors
        self.gamma = gamma  # Adjustment factor for boundary awareness
        self.random_state = random_state

    def fit_resample(self, X, y):
        if self.random_state is not None:
            np.random.seed(self.random_state)

        # Identify minority and majority classes
        unique_classes, class_counts = np.unique(y, return_counts=True)
        minority_class = unique_classes[np.argmin(class_counts)]
        majority_class = unique_classes[np.argmax(class_counts)]

        minority_indices = np.where(y == minority_class)[0]
        majority_indices = np.where(y == majority_class)[0]

        X_minority = X[minority_indices]
        X_majority = X[majority_indices]

        n_minority = len(X_minority)
        n_majority = len(X_majority)
        n_synthetic = n_majority - n_minority  # Number of samples to generate

        if n_synthetic <= 0:
            return X, y

        # Find safe minority samples (far from majority class)
        safe_minority_indices = self._find_safe_minority_samples(X_minority, X_majority)

        synthetic_samples = []

        for _ in range(n_synthetic):
            if len(safe_minority_indices) < 2:
                # If not enough safe samples, use all minority samples
                safe_minority_indices = list(range(len(X_minority)))

            # Step 1: Base Sample Generation
            idx1, idx2 = np.random.choice(safe_minority_indices, 2, replace=False)
            x1, x2 = X_minority[idx1], X_minority[idx2]

            delta = np.random.random()
            x_base = x1 + delta * (x2 - x1)

            # Step 2: Boundary Vector Calculation
            x_maj = self._find_nearest_majority_sample(x_base, X_majority)
            v = x_base - x_maj
            v_norm = v / (np.linalg.norm(v) + 1e-8)  # Normalize with epsilon

            # Step 3: Final Sample Adjustment
            x_synthetic = x_base + self.gamma * v_norm

            synthetic_samples.append(x_synthetic)

        X_resampled = np.vstack([X, synthetic_samples])
        y_resampled = np.hstack([y, np.full(len(synthetic_samples), minority_class)])

        return X_resampled, y_resampled

    def _find_safe_minority_samples(self, X_minority, X_majority):
        """Find minority samples that are far from majority class"""
        if len(X_majority) == 0:
            return list(range(len(X_minority)))

        # Calculate distances from each minority sample to all majority samples
        distances = []
        for x_min in X_minority:
            min_dist = np.min(np.linalg.norm(X_majority - x_min, axis=1))
            distances.append(min_dist)

        # Select top 50% of samples with largest distances (safest samples)
        n_safe = max(2, len(X_minority) // 2)
        safe_indices = np.argsort(distances)[-n_safe:]

        return safe_indices.tolist()

    def _find_nearest_majority_sample(self, x, X_majority):
        """Find the nearest majority class sample to x"""
        if len(X_majority) == 0:
            return x

        distances = np.linalg.norm(X_majority - x, axis=1)
        nearest_idx = np.argmin(distances)
        return X_majority[nearest_idx]

# ----------------------------------------------------------------------------
```

```python
# Function for generating Keras model
# ----------------------------------------------------------------------------
def generateModel(num_hidden_layers=1, num_neurons= 32, regularization="l1", lambda_value=1, num_features=1, num_classes=1):
  model = keras.Sequential()

  for i in np.arange(1,num_hidden_layers+1):
    # Regularization
    if (regularization=="l1"):
      kernel_regularizer=keras.regularizers.L1(lambda_value)
    else:
      kernel_regularizer=keras.regularizers.L2(lambda_value)

    # First hidden layer includes features input
    if (i == 1):
      model.add(keras.layers.Dense(units=num_neurons, input_shape=(num_features,), kernel_regularizer=kernel_regularizer))
    else:
      model.add(keras.layers.Dense(units=num_neurons, kernel_regularizer=kernel_regularizer))

    model.add(keras.layers.BatchNormalization())
    model.add(keras.layers.LeakyReLU(alpha=0.1))
    model.add(keras.layers.Dropout(0.01))

  # Output layer
  model.add(keras.layers.Dense(units=num_classes, activation="softmax", kernel_regularizer=kernel_regularizer))

  # plot_model(model, show_shapes=True, show_layer_names=True)

  # Return model
  return model

# ----------------------------------------------------------------------------
```

```python
# Define model and pipeline
# Return tuned pipeline
# ----------------------------------------------------------------------------
def simulate(x_train, y_train, num_classes, baws=False, hidden_layers_value=None, lambda_value=None):
  # General random seeds for experiment reproducibility
  np.random.seed(1)
  tf.random.set_seed(1)
  tf.keras.utils.set_random_seed(1)

  # Hyperaparameter and fixed values
  if hidden_layers_value is None:
    num_hidden_layers = [1, 3, 5]
  else:
    num_hidden_layers = [hidden_layers_value]

  num_neurons = [2048]
  epochs = [10]
  batchs = [32]
  regularization = ["l2"]

  if lambda_value is None:
    lambdas = [1e-6, 1e-3, 1e-1]
  else:
    lambdas = [lambda_value]

  # Keras wrapper
  clf = KerasClassifier(
    model = generateModel,
    loss= "sparse_categorical_crossentropy",
    metrics= ["acc"],
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    batch_size=batchs,
    epochs=epochs,
    num_hidden_layers = num_hidden_layers,
    num_neurons = num_neurons,
    regularization = regularization,
    lambda_value = lambdas,
    num_features=x_train.shape[1],
    num_classes=num_classes,
  )

  if baws:
    pipeline = imbPipeline([
        ("baws", BAWS(random_state=42, gamma=0.1)),
        ("scaler", StandardScaler()),
        ("clf", clf)
    ])
  else:
    pipeline = skPipeline([
        ("scaler", StandardScaler()),
        ("clf", clf)
    ])

  # Repeated kfold cross validation definition
  rkfold = RepeatedStratifiedKFold(n_splits=5, n_repeats=2, random_state=45)

  # Parameters of grid
  param_grid = {
                "clf__num_hidden_layers": num_hidden_layers,
                "clf__num_neurons": num_neurons,
                "clf__epochs":epochs,
                "clf__batch_size":batchs,
                "clf__regularization": regularization,
                "clf__lambda_value": lambdas,
              }

  # GridSearchCV
  search = GridSearchCV(pipeline, param_grid, cv=rkfold, verbose=4)

  result = search.fit(x_train, y_train)

  # Training with best hyperparameters
  pipeline.set_params(
    clf__num_hidden_layers=result.best_params_["clf__num_hidden_layers"],
    clf__num_neurons = result.best_params_["clf__num_neurons"],
    clf__epochs= result.best_params_["clf__epochs"],
    clf__batch_size= result.best_params_["clf__batch_size"],
    clf__regularization=result.best_params_["clf__regularization"],
    clf__lambda_value= result.best_params_["clf__lambda_value"]
    )

  print(f"Best result at validation set: {result.best_score_} with hyperparameters {result.best_params_}")

  print("Retraining:")
  pipeline.fit(x_train, y_train)
  return pipeline

# ----------------------------------------------------------------------------
# Present results
# Present results on screen
# ----------------------------------------------------------------------------
def visualizeResults(x_test, y_test, pipeline):
  # Test evaluation
  y_test_pred_proba = pipeline.predict_proba(x_test)

  y_test_pred = pipeline.predict(x_test)
  f1score = mt.f1_score(y_test, y_test_pred, average="weighted")
  acc = mt.accuracy_score(y_test, y_test_pred)
  preci = mt.precision_score(y_test, y_test_pred, average="weighted")
  recall = mt.recall_score(y_test, y_test_pred, average="weighted")

  # Classification report
  print("Classification report")
  report = mt.classification_report(y_test, y_test_pred)
  print(report)

  # Confusion matrix
  # TN FP
  # FN TP
  # print("Confusion matrix")
  # print(y_test_pred)
  labels = [
    "Benign",
    "Spyware",
    "Ransomware",
    "Trojan Horse"
  ]

  cm = mt.confusion_matrix(y_test, y_test_pred)
  pd_cm = pd.DataFrame(cm,
             columns=labels,
             index=labels)
  # print(pd_cm)

  # Display confusion matrix using ConfusionMatrixDisplay
  # g_cm = mt.ConfusionMatrixDisplay(cm,display_labels=labels).plot()
  # plt.show(g_cm)

  # Display confusion matrix using Heat Map
  ax= plt.subplot()
  sns.heatmap(cm, annot=True, fmt='g', ax=ax,
    square=True,
    cmap="Blues",
    linewidths=0.2,
    annot_kws={"fontsize":10.5})
  ax.set_xlabel("Predicted values", fontsize=8)
  ax.set_ylabel("True values", fontsize=8)
  ax.xaxis.set_ticklabels(labels, fontsize=8)
  ax.yaxis.set_ticklabels(labels, fontsize=8)
  cbar = ax.collections[0].colorbar
  cbar.ax.tick_params(labelsize=8)
  plt.yticks(rotation=0)
  plt.xticks(rotation=90)

  # ROC curves
  import colorcet as cc

  palette = sns.color_palette(cc.glasbey, 16)
  palette = [palette[10],palette[0]]+palette[1:10]+palette[11:]

  t_test_oneh = to_categorical(y_test,4)
  fig, ax = plt.subplots()
  auc = 0

  for i in np.arange(0,4):
    fpr, tpr, thresholds = mt.roc_curve(t_test_oneh[:, i], y_test_pred_proba[:, i])
    auc_value = round(mt.auc(fpr, tpr),4)
    auc = auc+auc_value
    plt.plot(fpr, tpr,color=palette[i],label=labels[i] + " (AUC="+str(auc_value)+")")
    #  Roc curve display alternative
    #  mt.RocCurveDisplay.from_predictions(
    #     t_test_oneh[:, i],
    #     y_test_pred_proba[:, i],
    #     name=f"{labels[i]}",
    #     color=palette[i],
    #     ax=ax
    # )

  plt.plot([0,1],[0,1],"k--")
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.legend(loc="best", prop={'size': 6})
  plt.grid()
  plt.show()

  auc = auc/4

  # Test evaluation results
  print("Test evaluation results: recall =", recall, ", F1 =", f1score, ", acc =", acc, ", precision =", preci, "auc =", auc)
```

```python
# Build model with BAWS on training data

# Check for and handle NaN values in x_train before passing to simulate
if np.isnan(x_train).any():
    print("NaN values detected in x_train. Imputing with column mean.")
    # Use pandas DataFrame for easier imputation, then convert back to numpy
    x_train_df = pd.DataFrame(x_train)
    x_train_df = x_train_df.fillna(x_train_df.mean())
    x_train = x_train_df.to_numpy()

# Also check x_test, though not strictly causing the current error, it's good practice for consistency
if np.isnan(x_test).any():
    print("NaN values detected in x_test. Imputing with column mean (from x_train).")
    # Apply the same mean imputation as training data to avoid data leakage
    x_test_df = pd.DataFrame(x_test)
    # Use the mean computed from the training data (x_train_df.mean()) for x_test imputation
    x_test_df = x_test_df.fillna(x_train_df.mean())
    x_test = x_test_df.to_numpy()

pipeline = simulate(x_train, y_train, num_classes, baws=True)

# Visualize results obtained
print("----------------------------------------------")
print("4-Classes Classifier results with BAWS")
print("----------------------------------------------")
visualizeResults(x_test, y_test, pipeline)
```
